```python
import numpy as np
import time
import random
from colorama import Fore, Back, Style
```

```python
c_barco = 'O'
c_tocado = 'X'
c_agua = ' '
c_agua_tiro = '/'
c_niebla = '~'
turnos = 0
hay_ganadore = 0
break_point = 0
id_jugador = ''
filas = 10
columnas = 10
disparo_jugador_x = []
disparo_jugador_y = []
barcos_jugador = {1: 4, 2: 3, 3: 2, 4: 1}
barcos_maquina = {1: 4, 2: 3, 3: 2, 4: 1}
tablero_niebla = np.full((filas, columnas), c_niebla)
tablero_maquina = np.full((filas, columnas), c_agua)
tablero_jugador = np.full((filas, columnas), c_agua)
```

```python
def bienvenida():
    global id_jugador
    print(f"{Fore.CYAN}Bienvenid@ al juego de 'Hundir la Flota'{Style.RESET_ALL}")
    id_jugador = input(f"¬øC√≥mo te gustar√≠a que te llamara?")
    time.sleep(1)
    print(f"{Fore.CYAN}Encantado de conocerte: \U0001F99A{Style.BRIGHT}{Fore.WHITE}{Back.BLACK}\033[4m{id_jugador}\033[0m{Style.RESET_ALL}, {Fore.CYAN}soy tu oponente: \U0001F916{Fore.WHITE}{Back.BLACK}\033[4mShipBrainy\033[0m{Style.RESET_ALL}{Fore.CYAN}, un experimentado estratega, asi que prep√°rate para perder...{Style.RESET_ALL}" )
    time.sleep(3)
    print(f"{Fore.CYAN}Estas son las instrucciones del juego:{Style.RESET_ALL}")
    time.sleep(2)
    print(f"""

      üè¥‚Äç‚ò†Ô∏è   {Fore.RED}{Fore.YELLOW}- Jugar√°s contra la m√°quina\n
      üè¥‚Äç‚ò†Ô∏è - En primer lugar, deber√°s colocar tus 10 barcos, con la siguiente distribuci√≥n:\n
                üö¢ - {Fore.GREEN}{Back.MAGENTA}4 barcos de 1 posici√≥n de eslora{Style.RESET_ALL}
                ‚õµ - {Fore.GREEN}{Back.MAGENTA}3 barcos de 2 posiciones de eslora{Style.RESET_ALL}
                ‚õµ - {Fore.GREEN}{Back.MAGENTA}2 barcos de 3 posiciones de eslora{Style.RESET_ALL}
                üö§ - {Fore.GREEN}{Back.MAGENTA}1 barco de 4 posiciones de eslora{Style.RESET_ALL}\n
      üè¥‚Äç‚ò†Ô∏è  {Fore.RED}{Fore.YELLOW}- Cuando comience el juego, tendr√°s un tablero para observar los disparos del contrincante, as√≠ como uno para comprobar tu jugada.\n
      üè¥‚Äç‚ò†Ô∏è - El juego se desarrolla por turnos, donde si alg√∫n jugador acierta durante el mismo, volver√° a repetir turno y en caso contrario, se producir√° un cambio de turno al oponente.\n
      üè¥‚Äç‚ò†Ô∏è - Para {Style.BRIGHT}\033[4mdisparar\033[0m{Style.RESET_ALL}{Fore.RED}{Fore.YELLOW} tendr√°s que introducir \033[4muna coordenada X(fila) y otra Y(columna)\033[0m{Fore.RED}{Fore.YELLOW}, con la cual marcar√°s d√≥nde quieres disparar en el tablero del oponente,
           y podr√°s ver los tableros nuevamente. En caso de introducir {Style.BRIGHT}\033[4mcoordenadas fuera del tablero\033[0m {Fore.RED}{Fore.YELLOW}, te saltar√° un error para que introduzcas coordenadas v√°lidas.{Style.RESET_ALL}\n
      üè¥‚Äç‚ò†Ô∏è {Fore.RED}{Fore.YELLOW}- El primer jugador que hunda todos los barcos del oponente, habr√° ganado, finalizando el juego.{Style.RESET_ALL}""")
    time.sleep(3)
    # print(f"\U0001F99A{Style.BRIGHT}{Fore.CYAN}{id_jugador}{Style.RESET_ALL}, {Fore.CYAN}¬øTienes las instrucciones claras(S/N)?{Style.RESET_ALL}")

    while True:
          respuesta = input(f"\U0001F99A{Style.BRIGHT}{Fore.WHITE}{Back.BLACK}\033[4m{id_jugador}\033[0m{Style.RESET_ALL},{Fore.CYAN}¬øTienes las instrucciones claras(S/N)?")
          if respuesta.upper() == "S":
               print(f"{Fore.CYAN}Comienza el ¬°¬°DESAF√çO!!{Style.RESET_ALL}")
               break
          elif respuesta.upper() =="N":
               print(f"{Fore.CYAN}Eres un poco {Style.BRIGHT}TORPE{Style.RESET_ALL}{Fore.CYAN}...üòàüòàüòà, asi que, l√©ete de nuevo las instrucciones que te he dado üòÇ")
               print(f"{Fore.CYAN}Si est√°s preparad@ pulsa: S, y si quieres comenzar de nuevo: N.")
               continue
          else:
               print(f"{Fore.CYAN}Error: ingresa la letra correcta: S o N{Style.RESET_ALL}")
               continue

    # print(f"{Fore.CYAN}¬øAlgo que decirme antes de comenzar?")
    input(f"{Fore.CYAN}¬øAlgo que decirme antes de comenzar?")
    print(f"{Fore.CYAN}Si,si...lo que tu digasüòÇ..¬°¬°PREP√ÅRATE!!üò†üò†...{Style.RESET_ALL}")
    time.sleep(2)
    print(f"{Fore.CYAN}Sabes que perder√°süòÇüòÇüòÇüòÇüòÇ, comienza el ¬°¬°DESAF√çO!!{Style.RESET_ALL} ")


def suspense(periodo = 0.3):
    for _ in range(5):
        time.sleep(periodo)
        print('.', end=' ')
    time.sleep(1)

def contador_turnos():
    global turnos
    turnos += 1

def mostrar_tableros():
    global turnos
    if turnos == 0:
        print('¬°Este es el campo de batalla!', end= '\n\n')
        time.sleep(1)
        print(f'Tu tablero:\n{tablero_jugador}\nTablero de ShipBrainy:\n{tablero_niebla} \n')
        time.sleep(3)
        print('Ahora, ¬°A los ca√±ones!', end= '\n\n')
    else:
        time.sleep(3)
        print(f'Tu tablero:\n{tablero_jugador}\nTablero de ShipBrainy:\n{tablero_niebla} \n')


def colocar_barco_aleatorio(tablero, tam_barco):
    while True:
        orientacion = np.random.choice(['horizontal', 'vertical'])
        fila = np.random.randint(0, filas)
        columna = np.random.randint(0, columnas)

        if orientacion == 'horizontal' and columna + tam_barco <= columnas:
            coordenadas = [(fila, c) for c in range(columna, columna + tam_barco)]
        elif orientacion == 'vertical' and fila + tam_barco <= filas:
            coordenadas = [(f, columna) for f in range(fila, fila + tam_barco)]
        else:
            continue

        ocupado = any(tablero[x][y] == c_barco for x, y in coordenadas if 0 <= x < filas and 0 <= y < columnas)
        ocupado_alrededor = any(tablero[x][y] == c_barco for x, y in get_celdas_alrededor(coordenadas) if 0 <= x < filas and 0 <= y < columnas)
        if not ocupado and not ocupado_alrededor:
            posicionar_barco(tablero, coordenadas)
            break

def get_celdas_alrededor(coordenadas):
    celdas_alrededor = set()
    for x, y in coordenadas:
        celdas_alrededor.update([(x-1, y), (x+1, y), (x, y-1), (x, y+1)])
    return celdas_alrededor

def posicionar_barco(tablero, coordenadas):
    for x, y in coordenadas:
        tablero[x][y] = c_barco

def colocar_barcos():
    global tablero_jugador
    for tam_barco, cantidad in barcos_jugador.items():
        for i in range(cantidad):
            colocar_barco_aleatorio(tablero_jugador, tam_barco)

def colocar_barcos_maquina():
    global tablero_maquina
    for tam_barco, cantidad in barcos_maquina.items():
        for i in range(cantidad):
            colocar_barco_aleatorio(tablero_maquina, tam_barco)

def test(coor1 = 1, coor2 = 2):
    global tablero_maquina
    tablero_maquina[coor1,coor2] = c_barco

def normal_test():
    prueba = input('¬øQuieres jugar el juego normal o una prueba?')
    if prueba.lower() == 'prueba':
      print('Vamos con el testeo ')
      test()
    else:
      colocar_barcos_maquina()

def obtener_coordenadas():
    global disparo_jugador_x
    global disparo_jugador_y
    global break_point
    while True:
        disparo_jugador_x = input("Ingresa la coordenada X (entre 0 y 9): ")
        if disparo_jugador_x.lower() == 'exit':  # Verifica si se ingresa 'exit' para salir del programa
           print("Has elegido salir del programa.")
           break_point += 1
           break

        disparo_jugador_y = input("Ingresa la coordenada Y (entre 0 y 9): ")
        if disparo_jugador_y.lower() == 'exit':  # Verifica si se ingresa 'exit' para salir del programa
           print("Has elegido salir del programa.")
           break_point += 1
           break
        try:
            disparo_jugador_x = int(disparo_jugador_x)
            disparo_jugador_y = int(disparo_jugador_y)

            if 0 <= disparo_jugador_x <= (filas-1) and 0 <= disparo_jugador_y <= (columnas-1):
                break
            else:
                print("Las coordenadas no est√°n en el rango v√°lido (0 a 9). Int√©ntalo de nuevo.")
        except ValueError:
            print("Ha ocurrido un error. Aseg√∫rate de ingresar n√∫meros enteros v√°lidos para las coordenadas.")

def disparo_jugador():
    ej = True
    contador = 0
    global break_point
    while ej:
        time.sleep(3)
        if contador != 0:
            print('¬°Te toca de nuevo! \n')
            mostrar_tableros()
        else:
            print('¬°Es tu turno!')
            time.sleep(1.5)

        obtener_coordenadas()
        if break_point != 0:
            break

        time.sleep(1)
        print(f"Disparo efectuado a las coordenadas ({disparo_jugador_x}, {disparo_jugador_y}).\n")
        suspense()

        if tablero_maquina[disparo_jugador_x, disparo_jugador_y] == c_barco:
            print('¬°Impacto! ¬°Barco alcanzado!\n')
            tablero_niebla[disparo_jugador_x, disparo_jugador_y] = c_tocado
            tablero_maquina[disparo_jugador_x, disparo_jugador_y] = c_tocado
            if condicion_victoria():
                break
            contador += 1
        elif tablero_maquina[disparo_jugador_x, disparo_jugador_y] == c_agua:
            print('Agua...\n')
            tablero_maquina[disparo_jugador_x, disparo_jugador_y] = c_agua_tiro
            tablero_niebla[disparo_jugador_x, disparo_jugador_y] = c_agua_tiro
            ej = False
        elif tablero_maquina[disparo_jugador_x, disparo_jugador_y] == c_tocado:
            print('Ya hab√≠as disparado ah√≠.\n')
            ej = False
        elif tablero_maquina[disparo_jugador_x, disparo_jugador_y] == c_agua_tiro:
            print('Agua, y ya hab√≠as disparado ah√≠.\n')
            ej = False

def disparo_maquina():
    ej = True
    contador = 0
    while ej:
        time.sleep(2)
        if contador != 0:
            print('¬°Le toca a ShipBrainy de nuevo! \n')
        else:
            print('¬°Es el turno de la m√°quina!')
        disparo_maquina_x = np.random.randint(0,10)
        disparo_maquina_y = np.random.randint(0,10)
        time.sleep(2)
        print(f"ShipBrainy dispara a las coordenadas ({disparo_maquina_x}, {disparo_maquina_y}).\n")
        suspense(0.5)
        if tablero_jugador[disparo_maquina_x, disparo_maquina_y] == c_barco:
            print('¬°Impacto!¬°Barco alcanzado!\n')
            tablero_jugador[disparo_maquina_x, disparo_maquina_y] = c_tocado
            if condicion_victoria():
                break
            contador += 1
        elif tablero_jugador[disparo_maquina_x, disparo_maquina_y] == c_agua:
            print('Agua...\n')
            tablero_jugador[disparo_maquina_x, disparo_maquina_y] = c_agua_tiro
            ej = False
        elif tablero_jugador[disparo_maquina_x, disparo_maquina_y] == c_tocado:
            print('Ya hab√≠as disparado ah√≠.\n')
            ej = False
        elif tablero_jugador[disparo_maquina_x, disparo_maquina_y] == c_agua_tiro:
            print('Agua, y ya hab√≠as disparado ah√≠.\n')
            ej = False

def condicion_victoria():
    global hay_ganadore
    ganador = False
    if (c_barco in tablero_jugador) == False:
        suspense()
        print(f'Lo siento {id_jugador}, gana ShipBrainy.')
        hay_ganadore +=1
        ganador = True
    elif (c_barco in tablero_maquina) == False:
        suspense()
        print(f'¬°Enhorabuena, {id_jugador}! ¬°Has ganado! \n Has tardado {turnos} turnos.')
        hay_ganadore +=1
        ganador = True
    return ganador
```

```python
bienvenida()
suspense()
colocar_barcos()
normal_test()
while True:
    mostrar_tableros()
    contador_turnos()
    disparo_jugador()
    if break_point != 0:
        print(f'¬°Te veo a la pr√≥xima, {id_jugador}!')
        break
    if hay_ganadore != 0:
        mostrar_tableros()
        break
    mostrar_tableros()
    disparo_maquina()
    if hay_ganadore != 0:
        mostrar_tableros()
        break
```
